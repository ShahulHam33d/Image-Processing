### Pre-processing of the image:

```python
from tensorflow.keras.preprocessing import image
img = image.load_img(filename,target_size=(224,224))
```

```python
resized_img = image.img_to_array(img)
print("image dimensions: ",resized_img.shape)
final_img = np.expand_dims(resized_img,axis=0) #we need 4 dimensions
final_img = tf.keras.applications.mobilenet.preprocess_input(final_img)
```

```python
final_img.shape
```

```python
predictions = mobile.predict(final_img)
#predictions
```

```python
from tensorflow.keras.applications import imagenet_utils

results = imagenet_utils.decode_predictions(predictions)
results
```

```python
#lets change the version of algo

mobiles = tf.keras.applications.mobilenet_v2.MobileNetV2
```

```python
pred = mobiles().predict(final_img)
result2 = imagenet_utils.decode_predictions(pred)
result2

```
